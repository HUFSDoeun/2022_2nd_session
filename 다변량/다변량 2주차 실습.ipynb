### (2) r 코드를 이요하여 

````r
X = matrix(c(rep(1,6), 0.5,6.0,1.5,1.1,2.0,0.5), ncol = 2, byrow = FALSE) # byrow = TRUE 옵션은 행방향으로 채워지게됨
X
````

````r
y = c(0.9,9.9,2.5,2.8,2.6,0.9)
y
````

````r
# inverse : solve, transpose : t , 행렬곱 : %*%
beta_hat = solve(t(X) %*% X) %*% t(X) %*% y # 단순 * 을 사용하게 되며 각 원소의 위치에 따라서 곱하기가 됨 -> 행렬곱 x
beta_hat
````
