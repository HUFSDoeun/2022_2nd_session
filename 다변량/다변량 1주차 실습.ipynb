확률 벡터 $\vec{x} = (x_1 x2)^T$에 대해 $x_1 \sim N_2(\vec{\mu}, \Sigma)$(책참고) 일때, x1의 확률 밀도 함수의 그래프는 아래와 같은  R코드를 작성 할 수 있음

```r
x1 = seq(-3,3, length = 1000)
y = dnorm(x1, mean = 0 , sd = 1)
plot (x1, y, type = "l", lwd = 1 , main = "pdf of x1") #lwd -> 선의 두께 옵션 
```

# 2.
이변량 정규분포의 확률 밀도 함수에 대한 등고선 그림과 3차원 그림은 R의 mvtnorm이라는 패키지의 dmvorm()함수를 이용하여 아래와 같이 구현한다.

- contour() : 등고선그림 그리기
- persp() : 3차원 투시도 그리기 

```r
# install.packages("mvtnorm")
library(mvtnorm)

x.points <-  seq(-3,3, length.out = 100) # x축
y.points = x.points # y축
z = matrix(0,nrow = 100, ncol = 100) #z축은 100x100의 매트릭스
mu = c(0,0) # 평균
sigma = matrix(c(2,1,1,1), nrow = 2) # 분산
for (i in 1:100){
  for (j in 1:100){
    z[i,j] = dmvnorm (c(x.points[i], y.points[j]),
                      mean = mu, sigma = sigma) # z의 높이는 정규분포를 따르고 이때 평균과 분산은 mu, sigma
  }
}
contour(x.points, y.points , z) # 등고선 그리기
```

```r
persp(x.points, y.points, z, phi = 45, theta = 30)
```

1번에서 정의된 결합확률밀도함수의 그림 그리기
## 1. 등고선 그림

```r
library(mvtnorm)
x.points = seq(-4, 4, length.out = 100)
y.points = x.points
z = matrix(0, nrow = 100, ncol = 100)
mu = c(0,1)
sigma = matrix(c(1,-1,-1, 2), nrow = 2)

for (i in 1:100){
  for (j in 1:100){
    z[i, j] = dmvnorm(c(x.points[i], y.points[j]),
                        mean = mu, sigma = sigma)
  }
}
```

```r
# 등고선
contour(x.points, y.points, z)
```

## 2. 3차원 투시도

```r
persp(x.points, y.points, z, phi = 45, theta = 30)
```

# 3
확률 변수 X가 이항분포 $ X \sim B(4,p)$를 따를때, 다음을 R을 이용해서 그려라.

## 1. p = 0.7일때 , X의 확률분포함수 막대그래프

```r
x = 0:4
x_prob = dbinom(x,4, 0.7) # x가 0~4 일때 성공확률이 0.7인 이항분포를 따른다면 각 값이 뽑힐 확률
x_prob
```

```r
sum(x_prob) # 다합하면 1
```

```r
# barplot그리기
barplot(x_prob, xlab = "x", ylab = "p(x)")
```

## 2. p 를 수평축으로 하고 P(X=1)을 수직축으로하는 그래프
각 p에 대해서 X=1일 확률에 대한 그래프

```r
p = seq(0,1, by = 0.1)
y = dbinom(x=1, size = 4, p = p)
plot(p, y, ylab = "P(X = 1")
```
